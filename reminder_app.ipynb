```python
import tkinter as tk
from tkinter import ttk  # Import ttk for themed tkinter
from tkinter import messagebox
import time
import threading
import pygame

# Initialize pygame mixer for sound
pygame.mixer.init()

# Flags to check if music is playing, if the popup is open, and if snooze is active
music_playing = False
popup_open = False
snooze_active = False

# Dictionary to map recurrence options to seconds
recurrence_options = {
    "Once": 0,
    "Daily": 24 * 3600,
    "Weekly": 7 * 24 * 3600,
    "Monthly": 30 * 24 * 3600,
}

# Function to remove snooze-related widgets
def remove_snooze_widgets():
    snooze_minutes_label.pack_forget()
    snooze_button.pack_forget()

def set_reminder():
    global music_playing, popup_open, snooze_active, reminder_time

    # Get the reminder message, time, and recurrence option from user input
    reminder_message = entry_message.get()
    hours = int(entry_hours.get())
    minutes = int(entry_minutes.get())
    seconds = int(entry_seconds.get())
    recurrence_option = recurrence_var.get()

    # Calculate the total time in seconds for the initial reminder
    reminder_time = (hours * 3600) + (minutes * 60) + seconds

    # Create a function to update the countdown timer
    def update_timer():
        global music_playing, popup_open, snooze_active, reminder_time
        while True:
            for i in range(reminder_time, -1, -1):
                hours, remainder = divmod(i, 3600)
                minutes, seconds = divmod(remainder, 60)
                countdown_var.set(f"Time Left: {hours:02d}:{minutes:02d}:{seconds:02d}")
                time.sleep(1)
            # When the timer reaches zero and snooze is not active, display the reminder message
            if not snooze_active and not popup_open:
                popup_open = True
                messagebox.showinfo("Reminder", reminder_message)
                if not music_playing:
                    music_playing = True
                    play_sound("See_You_Again.mp3")  # Change this to the path of your sound file
                    while pygame.mixer.music.get_busy():
                        if snooze_active:
                            break  # Exit the loop if snooze is activated
                        time.sleep(0.1)  # Check if the popup is still open
                    pygame.mixer.music.stop()
                    music_playing = False
                popup_open = False

            # Calculate the time for the next reminder based on recurrence option
            recurrence_time = recurrence_options.get(recurrence_option, 0)
            if recurrence_time == 0:
                break  # If recurrence is "Once," exit the loop
            time.sleep(recurrence_time)

    # Start the countdown timer in a separate thread
    countdown_thread = threading.Thread(target=update_timer)
    countdown_thread.start()

def play_sound(sound_file):
    try:
        pygame.mixer.music.load(sound_file)
        pygame.mixer.music.play()
    except Exception as e:
        messagebox.showerror("Error", f"Error playing sound: {str(e)}")

def on_popup_ok():
    global popup_open
    popup_open = False  # Set the popup_open flag to False when the "OK" button is clicked

def snooze_reminder():
    global snooze_active, reminder_time

    # Add 10 seconds to the reminder time
    reminder_time += 10

    # Set snooze_active to True
    snooze_active = True

    # Sleep for 10 seconds
    time.sleep(10)

    # Set snooze_active to False
    snooze_active = False

# Create the main window
root = tk.Tk()
root.title("Reminder App")

# Create a custom style for the countdown label with your preferred font color and style
custom_style = ttk.Style()
custom_style.configure("Countdown.TLabel", font=("Arial", 16), foreground="blue")  # Change font and color as desired

# Label and Entry for the reminder message
label_message = tk.Label(root, text="Reminder:")
label_message.pack()
entry_message = tk.Entry(root)
entry_message.pack()

# Label and Entry for the time input in hours
label_hours = tk.Label(root, text="Hours:")
label_hours.pack()
entry_hours = tk.Entry(root)
entry_hours.pack()

# Label and Entry for the time input in minutes
label_minutes = tk.Label(root, text="Minutes:")
label_minutes.pack()
entry_minutes = tk.Entry(root)
entry_minutes.pack()

# Label and Entry for the time input in seconds
label_seconds = tk.Label(root, text="Seconds:")
label_seconds.pack()
entry_seconds = tk.Entry(root)
entry_seconds.pack()

# Dropdown menu for recurrence options
recurrence_label = tk.Label(root, text="Recurrence:")
recurrence_label.pack()
recurrence_options = ["Once", "Daily", "Weekly", "Monthly"]
recurrence_var = tk.StringVar(root)
recurrence_var.set("Once")  # Default recurrence option
recurrence_menu = tk.OptionMenu(root, recurrence_var, *recurrence_options)
recurrence_menu.pack()

# Button to set the reminder
set_button = tk.Button(root, text="Set Reminder", command=set_reminder)
set_button.pack()

# Label to display the countdown timer with custom style
countdown_var = tk.StringVar()
countdown_label = ttk.Label(root, textvariable=countdown_var, style="Countdown.TLabel")
countdown_label.pack()

# Button to snooze the reminder
snooze_button = tk.Button(root, text="Snooze", command=snooze_reminder)
snooze_button.pack()

# Start the GUI event loop
root.mainloop()
```
